```python
import cv2
```

```python
img=cv2.imread("3 girs img.jpg")
cv2.imshow('shobi',img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cascade_data=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
faces=cascade_data.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30,30))
```

```python
for(x,y,w,h)in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    
```

```python
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
